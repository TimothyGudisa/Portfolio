## Side effects Classification

```python
# Create an empty dictionary to store side effect categories
side_effect_categories = {}

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    drug_name = row['drug_name']
    side_effects = row['side_effects']
    
    # Check if the side_effects column is empty or null
    if pd.isnull(side_effects) or side_effects == '':
        continue  # Skip this row if side_effects is empty or null
    
    # Split the side effects into individual words
    words = side_effects.split(', ')
    
    # Iterate over each word
    for word in words:
        word = word.strip()  # Remove leading/trailing spaces
        
        # Check if the word already exists in the dictionary
        if word in side_effect_categories:
            # Append the drug name to the existing word
            side_effect_categories[word].append(drug_name)
        else:
            # Create a new entry for the word
            side_effect_categories[word] = [drug_name]

# Sort the side effect categories by key
sorted_side_effect_categories = dict(sorted(side_effect_categories.items()))

# Print the categorized side effects in alphabetical order
for word, drugs in sorted_side_effect_categories.items():
    print( {word})
#     print(f"Drugs: {', '.join(drugs)}")
    print()
```

```python
import scikit_posthocs as sp
from scipy import stats

# Extract the columns needed for hypothesis testing
medical_condition = df_filtered['medical_condition']
side_effect_severity = df_filtered['severity_score_1']

# Calculate the frequency of incidence for each medical condition
freq_dict = {}
for condition in set(medical_condition):
    freq_dict[condition] = list(medical_condition).count(condition)

# Calculate the mean severity for each medical condition
mean_dict = {}
for condition in set(medical_condition):
    mean_dict[condition] = side_effect_severity[medical_condition == condition].mean()

# Calculate the standard deviation of severity for each medical condition
std_dict = {}
for condition in set(medical_condition):
    std_dict[condition] = side_effect_severity[medical_condition == condition].std()

# Perform Kruskal-Wallis H test between all medical conditions
test_stat, p_value = stats.kruskal(
    *[side_effect_severity[medical_condition == condition] for condition in set(medical_condition)]
)
print('Statistic:', test_stat, 'p-value:', p_value)

# Print the results
if p_value < 0.05:
    print('Reject null hypothesis: There is a significant difference in the side effect severity of medications for specific medical conditions.')
else:
    print('Fail to reject null hypothesis: There is no significant difference in the side effect severity of medications for specific medical conditions.')

# Perform Dunn's post-hoc test to determine which pairs of medical conditions are significantly different
posthoc_results = sp.posthoc_dunn(
    [side_effect_severity[medical_condition == condition] for condition in set(medical_condition)],
    p_adjust='bonferroni'
)

print("\nDunn's test results:")
print(posthoc_results)
```
